```python
# Finds the correlation of three catagories
import seaborn as sns
carData = pd.read_csv('cars1.csv')
plt.plot(carData['mpg'], 'o',  color='red')
plt.plot(carData['cylinders'], 'o', color='blue')
plt.plot(carData['weight']/46, 'o', color='green')

plt.legend(['mpg', 'cylinder', 'weight'])
plt.title("Correlation Between mpg, cylinder and weight")
plt.show()

#Prints Correlation betweena all three
correlation = carData[['mpg', 'weight', 'cylinders']].corr()
print(correlation)
```
